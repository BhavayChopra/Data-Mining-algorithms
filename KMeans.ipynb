```python
doc=open("content_words1.txt")
words=doc.read().lower().split()
print("The content words are: \n",words)
doc2_temp=open("dataset1.txt")
docs_temp=doc2_temp.read().lower().split("\n")
docs=[]
for i in range(1,len(docs_temp),2):
 docs.append(docs_temp[i])
print("\nThe document dataset is: \n",docs)
import re
count=0;
count_vectors=[]
for i in range(0,len(docs)):
 temp=[]
 for j in range(0,len(words)):
  m=re.findall(words[j],docs[i])
  temp.append(len(m))
 count_vectors.append(temp)
print("\nThe vector list is: ")
for i in range(0,len(count_vectors)):
 for j in range(0,len(count_vectors[i])):
  print(count_vectors[i][j],end=" ")
 print(" ")
import math
def dis(x,y):
 temp=0
 for i in range(len(x)):
  temp=temp+math.pow(x[i]-y[i],2)
 return round(math.sqrt(temp),2)
euc_dis=[]
count=0
for i in range(0,len(count_vectors)):
 temp1=[]
 for j in range(0,len(count_vectors)):
  temp1.append(dis(count_vectors[i],count_vectors[j]))
 euc_dis.append(temp1)
print("\n\nThe Euclidean distance is: ")
for i in range(0,len(euc_dis)):
 for j in range(0,len(euc_dis[i])):
  print(euc_dis[i][j],end=" ")
 print(" ")
def centroid(x):
 centroids=[]
 for i in range(0,len(x)):
  if(len(x[i])!=1):
   centr=[]
   for j in range(0,len(x[i])):
    for k in range(0,len(x[i][j])):
     if j==0:
      centr.append(x[i][j][k])
     else:
      centr[k]=centr[k]+x[i][j][k]
   for j in range(0,len(centr)):
    centr[j]=round(centr[j]/len(x[i]),2)
   centroids.append(centr)
  else:
   centroids.append(x[i][0])
 return centroids
def check(x,y):
 count=0
 for i in range(0,len(x)):
  if(x[i]==y[i]):
   count=count+1
 if(count==len(x)):
  return True
 else:
  return False
def k_means_clustering(kvalues,count_vectors):
 clu=[]
 clu_num=[]
 centroids1=[]
 centroid_final=[]
 for i in kvalues:
  centroid_final.append(count_vectors[i])
  if(i!=max(kvalues)):
   centroids1.append(count_vectors[i+1])
  else:
   centroids1.append(count_vectors[0])
 count=0
 while(check(centroids1,centroid_final)==False):
  print("\nThe centroid for iteration ",count,"is:\n",centroid_final)
  centroids1=centroid_final
  clu=[]
  clu_num=[]
  for i in range(0,len(centroid_final)):
   clu.append([])
   clu_num.append([])
  for i in range(0,len(euc_dis)):
   distance=[];
   for j in range(0,len(centroid_final)):
    distance.append(dis(count_vectors[i],centroid_final[j]))
   minvalue=0;
 
   count4=0
   for k in range(0,len(kvalues)):
    if(i==k):
     count4=count4+1
   for j in range(0,len(centroid_final)):
    if(distance[j]<=distance[minvalue] and count4==0):
     minvalue=j
   if(count4!=0):
    clu[i].append(count_vectors[i])
    clu_num[i].append(i)
   else:
    clu[minvalue].append(count_vectors[i])
    clu_num[minvalue].append(i)
  print("\nThe cluster for iteration ",count,"is:\n",clu,"\n\n")
  centroid_final=centroid(clu)
 
  count=count+1
 
 print("\nTherefore the clusters formed are ")
 print(clu_num)
noofkvalues=int(input("\nEnter the no of kvalues : "))
kvalues=[]
for i in range(0,noofkvalues):
 kvalues.append(i)
print("\n",kvalues)
k_means_clustering(kvalues,count_vectors)
```
